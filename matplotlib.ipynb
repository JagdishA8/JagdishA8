```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,2,1)
plt.plot(x,y_sin)
plt.title('sin axis')

plt.subplot(2,2,2)
plt.plot(x,y_cos)
plt.title('cos axis')
```

```python
x=[np.random.normal(1,std,100) for std in range(1,4)]
plt.boxplot(x,vert=True,patch_artist=True)
plt.show()
```

```python
outlier=[]
def detect_outlier(data):
    threshold=2
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)<threshold:
            outlier.append(i)
        
detect_outlier(x)    
```
